### ● 確率変数

演習１の $x$ の値は，実行ごとに異なり，次の $x$ は予測困難です．
<br/>このようなランダムな変数 $x$ のことを，**確率変数**または**ランダム変数 (random variable)** といいます．

このような値が定まらない変数（確率変数）を，数学的にどのように定義すべきか，長年の謎でした．
<br/>この謎は，現代確率論 (Kolmogorov, 1933) によって，次のように回避されました．

- 確率変数 $x$ とは，識別子 $\omega$ の関数 $x[\omega]$ のことである！

#### >>>> dict 型による実装

次のコードは，Python の dict 型による，確率変数のプログラミング例です．
<br/> dict 型は，適当な識別子（`key`という）に紐づけた値を呼び出すデータ構造です．

````python
x = {
    'a':np.random.uniform(),
    'b':np.random.uniform(),
    'c':np.random.uniform()
}

print(x)
````

この `x` に識別子 $\omega=$ `'a'`, `'b'`, `'c'` を与えると，対応する値が選ばれます．

````python
omega = 'a'; print(x[omega])
omega = 'b'; print(x[omega])
omega = 'c'; print(x[omega])
````

このような，識別子 $\omega$ に対応する数値を記憶させた関数 $x[\omega]$ のことを，**確率変数**と呼ぶわけです．

以上の例では，識別子を3個としましたが，$x$ の全ての値を網羅するには足りません．
<br/>そこで，可能な全ての識別子を収めた巨大な集合 $\Omega$ を考え，それを**標本空間 (sample space)** と呼びます．
<br/>また，個々の識別子 $\omega$ を，**標本点 (sample point)** と呼びます．

#### >>>> リスト型による実装

確率変数 $x(\omega)$ 風を醸し出すために dict 型を使いましたが，考えてみればリスト型でも同じことです．識別子が整数になるだけです．

````python
x = [
    np.random.uniform(),
    np.random.uniform(),
    np.random.uniform()
]
print(x)

omega = 0; print(x[omega])
omega = 1; print(x[omega])
omega = 2; print(x[omega])
````

### ● 標本と母集団

その他，確率論の前身となった**統計学 (statistics)** の用語も導入しておきます．これらもよく使います．

- **試行 (trial)** ・・・ 結果を得るための操作を，**試行**といいます．演習１の例では，セルの実行が試行になります．
- **試行結果 (trial result)** ・・・ 試行の結果を，**試行結果**といいます．演習１の例では，出てきた具体的な数が試行結果になります．
- **標本 (sample)** ・・・ 具体的な試行結果を集めた集合を，**標本**といいます．
- **標本値 (sample value)** ・・・ 標本の要素（標本に含まれる試行結果）を，**標本値**といいます．
- **標本数 (sample number)** ・・・ 標本の要素数を，**標本数**といいます．
- **母集団 (population)** ・・・ 起こり得る全ての試行結果を集めた，理想的な集合を，**母集団**といいます．
- **サンプリング (sampling)** ・・・ 標本を抽出すること．概念的には，母集団から抽出すると考える．

※「標本」を「サンプル」に言い換えても，正式な専門用語です．（サンプル，サンプル数，…）


紛らわしい用語があるので要注意です．

- **標本点 (sample point)** ・・・ 標本の要素（数値）のことではありません．標本空間の識別子 $\omega$ （試行結果の種類）のことです．

### ● 標本空間，母集団，標本

統計学の母集団と，確率論の標本空間は，互いに似た概念ですが，次のような違いがあります．

- **標本空間 $\Omega$** ＝ 母集団から試行結果の重複を除いたもの（の識別子）．

ようするに，標本空間とは，起こり得る全ての試行結果の**種類**を納めた集合で，算数の数直線のような役割です．数直線上の異なる点は，全て異なる数であり，同じ数が横並びしたりはしません．それと似たような理想化概念です．

例えば，サイコロ投げの標本空間 $\Omega$ は重複の無い 1～6 の集合ですが，
<br>母集団（＝無限試行）には，同じ数字が無限個含まれます．

|標本空間|母集団|標本|
|:--|:--|:--|
|Ω = {1, 2, 3, 4, 5, 6}|U = {2, 1, 3, 6, 6, 1, 4, 1, 2, 4, 4, 5, 4, 6, 4, 4, 2, 5, 1, 5, ...}|Π = {6, 2, 4, 1, 2, 4, 3, 5}|
|試行結果の種類|試行結果の全体|試行結果の一部（母集団の部分集合）|

## ■ 疑似乱数

以上に見たランダムな数は，実際には，**疑似乱数**と呼ばれるものです．
<br/>ランダム現象の数値計算で不可欠なものなので，説明します．

**乱数 (random number)** とは，無作為に定まる予測不可能な数のことです．2種に大別されます．



1.   真の乱数 (true random number, TRN) ... 物理現象を利用して生成される完全に予測不可能な数．
2.   疑似乱数 (pseudo-random number, PRN) ... 計算によって生成されるランダムに見える数列．



改めて疑似乱数の標本を見てみます．


````python
np.random.seed(1) #シードを 1 に設定する．
np.random.uniform(size=5) #一様乱数（一様に分布する乱数）
````

- このような疑似乱数は，何らかの漸化式 $a_{i+1}=f(a_{i})$ によって生成されています．
- このようにして，順番に生成される乱数の列を，**乱数列 (random sequence)** といいます．
  - 乱数列は「標本」の一種ですが，一般の標本とは異なり，要素に（時間的な）順番があります．
  - これまでの例では，乱数列の「標本の側面」に着目して，利用してきたわけです．
- この漸化式の内部的な初期値に相当するものを，**シード (seed)** といいます．
- 同じシードに対しては，同じ乱数列が何度でも生成されます．
- `np.random.seed(1)` を省いたときのシードは，Numpy 内部で無作為に自動設定されます．


````python
np.random.seed(1)
np.random.uniform(size=5)
````

シードを変えると，また別の乱数列が生成されます．

````python
np.random.seed(2) #シードを 2 に設定する．
np.random.uniform(size=5)
````

### ● 疑似乱数の性質

1. 同じシードに対して，同じ乱数列が再現してしまうので，疑似乱数は，真の乱数ではありません．
2. ただし，優秀な乱数生成アルゴリズムは，驚異的に長い周期を有しています（＝ 同じパターンが再帰するまでの回数が膨大）
3. いま使用している乱数生成アルゴリズムは，次で確認できます．

````python
np.random.default_rng()
````

- Numpyのデフォルトでは PCG-64 という乱数生成アルゴリズムが使われるようです．
- PCG-64 は，$2^{128} \approx 3.4\times 10^{38}$ という極めて長い周期を有します．<br/>[Permuted congruential generator (64-bit, PCG64) — NumPy v2.3 Manual](https://numpy.org/doc/stable/reference/random/bit_generators/pcg64.html)
- PCG-64 の先代にあたるのが，日本人数学者による [メルセンヌ・ツイスタ - Wikipedia](https://ja.wikipedia.org/wiki/%E3%83%A1%E3%83%AB%E3%82%BB%E3%83%B3%E3%83%8C%E3%83%BB%E3%83%84%E3%82%A4%E3%82%B9%E3%82%BF) で，長らく世界を席巻していました．現時点では，これによる実装も多く残存していますが，より高速な PCG-64 への置き換えが進みそうです．その途上でさらに優秀なアルゴルズムが発明される可能性もあります．

## ■ ヒストグラムと確率密度関数


一般に，ランダムに出現する値の頻度を表した数表や数式のことを，**確率分布 (probability distribution)** といいます．

- 確率分布の数表をグラフ化したものを，**ヒストグラム (histogram)** といいます．
- ヒストグラムを理想化して，数式化したものを，**確率密度関数 (probability density function, PDF)** といいます．

### ● ヒストグラム

演習４で見た乱数は，実は，**一様乱数 (uniform random number)** という乱数でした．一様乱数とは，区間[0,1)上に均等に分布する乱数のことです．
これを具体的に，**ヒストグラム**で見てみましょう．

サンプル数1000の一様乱数のヒストグラムを以下に示します．

````python
# np.random.seed(2) #シードを 2 に設定する．
x_sample = np.random.uniform(size=1000)
plt.hist(x_sample, bins=20, histtype="bar", edgecolor='black')
plt.xlabel('x')
plt.ylabel('Frequency')
````

ヒストグラムの見方は次の通りです．

1. 棒（長方形）の底辺が表す区間を，**ビン (bin)** または**階級** と呼びます．
2. ビンの幅を，**ビン幅 (bin width)** または**階級幅**といいます．
3. 棒の高さを，**度数 (frequency)** と呼び，各ビンに入った標本点の個数を表します．

> ### 演習５
>
> 1. シードを指定せずに実行すると，ヒストグラムが変化することを確かめよ．
> 2. 標本数を増やしながら，ヒストグラムの本来の形状を推定せよ．



（注：一様乱数のヒストグラムは，理論上は一定値になるはずですが，有限の標本数ではなりません）

### ● 確率密度関数 $p(x)$

演習５で見たように，ヒストグラムの高さは，サンプル数に依存します（多数なほど高い）．そのままでは，サンプル数の異なる分布を比較しにくいので，**確率密度関数(probability density function, PDF)** を導入します．

ここで，**確率密度関数**とは，
- **棒（長方形）の面積が**，各ビンに入る**確率に一致**するよう，ヒストグラムの高さを調整し，
- それを **$x$ の関数 $p(x)$ として表示したもの**

です．全てのビンのどれかに入る確率は $1$（100%）なので，必然的に次の性質が要請されます．

> **公理（全確率 $1$）：**
>
> - 確率密度関数 $p(x)$ の総面積，すなわち全確率は，必ず $1$ になる．
>
> - すなわち $\displaystyle \int_{\infty}^{\infty} p(x) dx = 1$．

というわけで，棒の面積の総和が $1$ になるように高さ調整します．具体的には，度数（もとの高さ）を標本数で割り，さらにビン幅で割ればよいです．

- `plt.hist`関数では，オプション引数 `density=True` を追加すると，自動的にそうしてくれます．

````python
sample_x = np.random.uniform(size=5000000)
plt.hist(sample_x, bins=20, histtype="bar", edgecolor='black', density=True)
plt.xlabel('$x$')
plt.ylabel('$p(x)$')
````

### 確率密度関数に関する注意

- 上で求めた確率密度関数は，厳密にいうと，近似です．
- 正式には，標本数 $\to\infty$ で浮かび上がる**理想的な関数形** $p(x)$を，**確率密度関数**と呼びます．

## ■ 様々な分布

> ### 演習６
>
> 以下，代表的な確率分布に従う乱数を生成し，それらの確率分布を観察してみます．サンプル数が足りない例を示すので，各自サンプル数を増やしながら，理想的な関数形を想像してみてください．



### ● 区間 $a\leq x <b$ 上の一様分布 (uniform destribution)

````python
# np.random.seed(1)
sample_x = np.random.uniform(high=1, low=2, size=1000)
plt.hist(sample_x, bins=20, histtype="bar", edgecolor='black', density=True)
plt.xlabel('$x$')
plt.ylabel('$p(x)$')
````

### ● 正規分布 (normal destribution) $N(m,\sigma)$　※平均$m$，標準偏差$\sigma$

別名，**ガウス分布 (Gaussian destribution)** ともいいます．物理学方面ではガウスのほうが主流かも．

> Gaussian は人名由来なので常に大文字です．

````python
sample_x = np.random.normal(loc=1, scale=2, size=1000)	# 平均1，標準偏差2の正規分布
plt.hist(sample_x, bins=20, histtype="bar", edgecolor='black', density=True)
plt.xlabel('$x$')
plt.ylabel('$p(x)$')
````
