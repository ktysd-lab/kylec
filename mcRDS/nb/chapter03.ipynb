## ■ 確率過程とは

### ● 時系列

数列 $a[i]$ ($i=1,\cdots,n$) のなかで，
<br/>添字 $i$ が時間的順序を表すものを，**時系列 (time series)** といいます．
<br/>$n$ を，**時系列長 (time series length)** といいます．

### ● 確率過程

１回の試行によって，１本の時系列（振動波形とか）が得られるものを，**確率過程 (stochastic process)** といいます．
<br/>数学的には，「識別子 $\omega$」と「時系列の添字 $i$」の関数 $x[\omega,i]$ として表現されます．

確率過程 $x[\omega,i]$ に，具体的な識別子 $\omega$ を代入すると，個別の時系列 $x[i]$ が選び出されます．
<br/>選び出された個別の時系列を，**標本路 (sample path)** といいます．

> おみくじの箱に，振動波形がたくさん入っていて，それを引くイメージ．

逆に，具体的な $i$ を代入すると，その時刻の確率変数 $x[\omega]$ が選び出されます．

## ■ 時間統計とアンサンブル統計

確率過程 $x[\omega,i]$ は，標本点 $\omega$ と，時間 $i$ の２変数関数です．
<br/>２つある変数のどちらを固定するかで，２通りの解釈が可能です．

1. 確率過程 $x[\omega,i]$ とは，標本点 $\omega$ に標本路 $x[i]$ ($i=1,\cdots,n$) を紐づける確率変数のことである．
2. 確率過程 $x[\omega,i]$ とは，時刻 $i$ に依存した確率変数 $x[\omega]$ のことである．


これらに対応して，２種類の**統計 (statistics)** が存在します．

1. **時間統計 (time statistics)**：<br/>標本点 $\omega=\bar\omega$ に紐づいた標本路 $x[i]:=x[\bar\omega,i]$ を取り出す．その全ての項<br/>$\qquad
\{x[1],x[2],x[3],\cdots\}
:=
\{x[\bar\omega,1],x[\bar\omega,2],x[\bar\omega,3],\cdots\}
$<br/>を集計した統計のこと．
2. **アンサンブル統計 (ensemble statistics)**：<br/>時刻 $i=\bar i$ における確率変数 $x[\omega]:=x[\omega,\bar i]$ を取り出す．その全ての値<br/>$\qquad\{x[\omega_1],x[\omega_2],x[\omega_3],\cdots\}
:=
\{x[\omega_1,\bar i],x[\omega_2,\bar i],x[\omega_3,\bar i],\cdots\}
$<br/>を集計した統計のこと．

> **【定義の等号】** $A := B$ は「$A$ を $B$ で定義する」，$A =: B$ は 「$A$ で $B$ を定義する」を表します．

## ■ 標本路の時間統計


どちらも重要な統計ですが，時間統計から始めます．

ある標本点 $\bar\omega$ に紐づいた標本路 $x[i]:=x[\bar\omega,i]$ ($i=1,2,\cdots,n$) を考えます．

### ● 平均 (mean)と積率 (moment)：

次式を，**平均 (mean/average)** と呼びます．力学に例えると，**重心**に相当します．

$$
m_x
:=
\ave{x}
:=
\frac{1}{n}\sum_{i=1}^{n} x[i]
$$

> 各変位が $x[i]$ であるような，単位質点 $n$ 個の重心と，同じ計算式になっている．

標本点を明示したいときは $\ave{x[\bar\omega]}$ と書きましょう（$i$ は総和で消える）．

また，$n$乗の平均を，**$n$次モーメント (moment)** といいます．モーメントを**積率**ともいいます．

$$
m^{(n)}_x
:=
\ave{x^n}
:=
\frac{1}{n}\sum_{i=1}^{n} x[i]^n
$$

#### **》平均操作 $\ave{\cdot}$ の公式**

定数 $a$ と時系列 $x[i]$, $y[i]$ ($i=1,\cdots,n$) について，次の公式が成立します．

- $\ave{a}= a$
- $\ave{a x}= a\ave{x}$
- $\ave{x + y}=\ave{x}+\ave{y}$

> 証明） $\ave{a x} := \frac{1}{n}\sum_{i=1}^{n} ax[i] = a\frac{1}{n}\sum_{i=1}^{n} x[i] = a\ave{x}$ ※他も同様
>
> 計算例） $\ave{(x - m_x)^2} = \ave{x^2 - 2xm_x+ m_x^2} = \ave{x^2} - 2m_x\ave{x}+ m_x^2 = \ave{x^2} - m_x^2$

### ● 分散 (variance)：

次式を，**分散 (variance)** と呼び，分布の広がりの程度を表します．

$$
\sigma_x^2
:=
\ave{(x-m_x)^2}
:=
\frac{1}{n}\sum_{i=1}^{n} (x[i]-m_x)^2
$$

分散の正の平方根 $\sigma_x:=\sqrt{\sigma_x^2}$ を，**標準偏差 (standard deviation, SD)** といいます．



また，**分散**のことを，**中心 $2$ 次モーメント (2nd central moment)** ともいいます．

#### **》中心化 (centering)**

平均で引く操作 $x_c[i] := x[i] - m_x$ を，**中心化 (centering)** といいます．中心化後の平均は $0$ になります．

$$
\ave{x_c} = \ave{x - m_x} = \ave{x} - m_x = m_x - m_x = 0
$$

中心化は，平均値まわりの変動成分を抽出するのに役立ちます（特に変動が微小な場合）．分散は，中心化後の $2$ 次モーメントなので，**中心 $2$ 次モーメント**となります．


### ● 共分散 (covariance)：

ある標本点 $\bar\omega$ で選ばれた２種類の標本路 $x[i]$, $y[i]$ ($i=1,2,⋯,n$) に対して，次式を**共分散 (covariance)** といいます．

$$
\sigma_{xy}
:=
\langle (x-m_x)(y-m_y)\rangle
:=
\frac{1}{n}\sum_{i=1}^{n} (x[i]-m_x)(y[i]-m_y)
$$

> 「$\bar\omega=$ある人」から測定した「心電図の波形 $x[i]$」と「筋電図の波形 $y[i]$」などが相当します．

### ● 相関係数 (correlation coefficient)：

以下に述べる標準化後の共分散を，**相関係数 (correlation coefficient)** といいます．

$$
\rho_{xy}
:=
\left\langle \frac{(x-m_x)}{\sigma_x}\frac{(y-m_y)}{\sigma_y}\right\rangle
=
\frac{\sigma_{xy}}{\sigma_x\sigma_y}
$$

#### **》標準化 (standardization)**

中心化後に標準偏差で割る操作
$\displaystyle
x_s[i] := \frac{x[i]-m_x}{\sigma_x}
$ のことを，**標準化 (standerdization)** といいます．

標準化後の $x_s[i]$ は，平均 $0$，分散 $1$ になります．
平均 $0$ は中心化より明らか．分散 $1$ は次のように示されます．
$$
\left\langle \left(\frac{(x-m_x)}{\sigma_x}\right)^2\right\rangle
=
\frac{\ave{(x-m_x)^2}}{\sigma_x^2}
=
\frac{\sigma_x^2}{\sigma_x^2} = 1
$$

というわけで，**相関係数**は，標準化共分散とでも呼ぶべき統計量です（そんな用語はないかも）．具体的には，$x[i]$ と $y[i]$ の（平均，分散）を（$0$, $1$）に揃えてから計算した共分散のことです．

## ■ 自己相関関数

もう１つ，このあと重要になる時間統計があります．
**自己相関関数 (autocorrelation function, ACF)** といいます．

自己相関関数を使うと，標本路のランダム性の程度を評価できます．

### ● ランダム性の類似度解釈

自己相関関数のベースとなる考え方を導入します．

- $x[i]$ が完全にランダムであるとは，現在の値 $x[i]$ から，次の値 $x[i+1]$ を全く予測できないことである．

これを，次のように言い換えます．

- $x[i]$ が完全にランダムであるとは，次の２つの波形が，全く**類似しない**ことである．<br/>$
\begin{alignat}{7}
\quad&\text{時間をずらした波形:}\quad&&x[2],&\;&x[3],&\;&\cdots,&\;&x[i+1],&\;&\cdots,&\;&x[n]&\\
&\text{元の波形:}\quad&&x[1],&&x[2],&&\cdots,&&x[i],&&\cdots,&&x[n-1]&
\end{alignat}
$

### ● ベクトルの類似度

一般に，数の並びの類似度は，ベクトルの内積で評価できます．

$$
\bm{x}\cdot\bm{y}
:= \sum_{i=1}^{n} x_iy_i
= x_1y_1 + \cdots + x_ny_n
= \|\bm{x}\|\|\bm{y}\|\cos\theta
\quad\text{（ベクトルの内積）}
$$
$\|\bm{x}\|=\sqrt{x_1^2+\cdots+x_n^2}$ はベクトルの長さで，ノルムといいます．$\theta$ は２つのベクトルのなす角です．

この内積を，$\cos\theta$ について解いたものを，**コサイン類似度 (cosine similarity)** といいます．

$$
\mbox{Sim}(\bm{x}, \bm{y})
:=
\frac{\bm{x}\cdot\bm{y}}{\|\bm{x}\|\|\bm{y}\|}
=
\cos\theta
$$

これを用いて，ベクトル $\bm{x}$, $\bm{y}$ の類似度は，次のように評価されます．

|$\mbox{Sim}(\bm{x}, \bm{y})$|幾何学的関係&emsp;|類似度の判定|
|:---:|:---|:---|
|$+1$|$\bm{x}\mathop{/\!\!/}\bm{y}$ 平行|$\bm{x}$ と $\bm{y}$ は最も類似する|
|$0$|$\bm{x}\perp\bm{y}$ 直交|$\bm{x}$ と $\bm{y}$ は最も異なる|
|$-1$|$\bm{x}\mathop{/\!\!/}-\bm{y}$ 平行|$\bm{x}$ と $-\bm{y}$ は最も類似する|

同様の評価は，内積 $\bm{x}\cdot\bm{y}$ でも可能です．最大値がキリの良い$\pm 1$ではなくなりますが，0は保たれます．

|&emsp;&emsp;&emsp;&emsp;$\bm{x}\cdot\bm{y}$&emsp;&emsp;&emsp;|幾何学的関係&emsp;|類似度の判定|
|:---:|:---|:---|
|$+\max|\bm{x}\cdot\bm{y}|$|$\bm{x}\mathop{/\!\!/}\bm{y}$ 平行|$\bm{x}$ と $\bm{y}$ は最も類似する|
|$0$|$\bm{x}\perp\bm{y}$ 直交|$\bm{x}$ と $\bm{y}$ は最も異なる|
|$-\max|\bm{x}\cdot\bm{y}|$|$\bm{x}\mathop{/\!\!/}-\bm{y}$ 平行|$\bm{x}$ と $-\bm{y}$ は最も類似する|

### ● 時系列の類似度

#### **》時系列そのものの類似度**

次に，２つの時系列 $x[i], y[i]$ ($i=1,2,\cdots,n$)を考えます．
<br/>ここで，重要なアイデアとして，各項を成分とするベクトルを作ります．

$$
\bm{x} = (x[1], x[2], \cdots, x[n])
,\quad
\bm{y} = (y[1], y[2], \cdots, y[n])
$$

基本的には，これらの内積 $\bm{x}\cdot\bm{y}$ によって，時系列の類似度が評価できます．
<br/>評価されるのは，時系列の波形（グラフ形状）の類似度です．

#### **》時系列の変動成分の類似度 ＝ 共分散（の定数倍）**

次に，時系列の変動成分にフォーカスするため，時系列から平均を引きます（中心化します）．

$$
\begin{align}
 &\bm{x}_c = (x[1]-m_x, x[2]-m_x, \cdots, x[n]-m_x)
 ,\\
 &\bm{y}_c = (y[1]-m_y, y[2]-m_y, \cdots, y[n]-m_y)
\end{align}
$$

これらを，**平均中心化列 (mean-centered series)** といいます．


平均中心化後の内積を取り，時系列長で割ると，**共分散**に一致してしまいます．

$$
\frac{
    \bm{x}_c\cdot\bm{y}_c
}{
    n
}
=
\frac{1}{n} \sum_{i=0}^{n} (x[i] - m_x)(y[i] - m_y)
=
\sigma_{xy}
$$

($1/n$)の有無は類似度の評価結果を変えませんから，使い方は元の内積と同じです．

|&emsp;&emsp;&emsp;&emsp;共分散 $\sigma_{xy}$&emsp;&emsp;&emsp;|幾何学的関係&emsp;|類似度の判定|
|:---:|:---|:---|
|$+\max|\sigma_{xy}|$|$\bm{x}_c\mathop{/\!\!/}\bm{y}_c$ 平行|時系列 $x[i]$, $y[i]$ の変動成分が最も類似する|
|$0$|$\bm{x}_c\perp\bm{y}_c$ 直交|時系列 $x[i]$, $y[i]$ の変動成分が最も異なる|
|$-\max|\sigma_{xy}|$|$\bm{x}_c\mathop{/\!\!/}-\bm{y}_c$ 平行|時系列 $x[i]$, $-y[i]$ の変動成分が最も類似する|

#### **》時系列の変動成分のコサイン類似度 ＝ 相関係数**

同様に，平均中心化後のコサイン類似度を取ると，**相関係数**と完全に一致します．

$$
\begin{align}
    \mbox{Sim}(\bm{x_c}, \bm{y_c})
    &=
    \frac{
        \sum_{i=0}^{n} (x[i]-m_x) (y[i]-m_y)
    }{
        \sqrt{\sum_{i=0}^{n} (x[i]-m_x)^2}
        \sqrt{\sum_{i=0}^{n} (y[i]-m_y)^2}    
    }
    \\
    &=
    \frac{
        \frac{1}{n}\sum_{i=0}^{n} (x[i]-m_x) (y[i]-m_y)
    }{
        \sqrt{\frac{1}{n}\sum_{i=0}^{n} (x[i]-m_x)^2}
        \sqrt{\frac{1}{n}\sum_{i=0}^{n} (y[i]-m_y)^2}    
    }
    =
    \rho_{xy}
\end{align}
$$

このように，相関係数とは，コサイン類似度そのものなので，使い方も同じです．

|&emsp;&emsp;相関係数 $\rho_{xy}$&emsp;&emsp;|幾何学的関係&emsp;|類似度の判定|
|:---:|:---|:---|
|$+1$|$\bm{x}_c\mathop{/\!\!/}\bm{y}_c$ 平行|時系列 $x[i]$, $y[i]$ の変動成分が最も類似する|
|$0$|$\bm{x}_c\perp\bm{y}_c$ 直交|時系列 $x[i]$, $y[i]$ の変動成分が最も異なる|
|$-1$|$\bm{x}_c\mathop{/\!\!/}-\bm{y}_c$ 平行|時系列 $x[i]$, $-y[i]$ の変動成分が最も類似する|

### ● 自分との類似度

冒頭で述べた**ランダム性の類似度解釈**を復習します．

- $x[i]$ が完全にランダムであるとは，次の波形が全く**類似しない**ことである．<br/>$
\begin{alignat}{7}
\quad&\text{元の波形:}\quad&&x[1],&&x[2],&&\cdots,&&x[i],&&\cdots,&&x[n-1]&
\\
\quad&\text{時間をずらした波形:}\quad&&x[2],&\;&x[3],&\;&\cdots,&\;&x[i+1],&\;&\cdots,&\;&x[n]&
\end{alignat}
$

この主張にある**類似度**を，次のベクトルの内積 $\bm{x}_{i}\cdot\bm{x}_{i+1}$で評価していきます．

$$
\begin{alignat}{7}
&\bm{x}_{i}&&:=\big(x[1],&&x[2],&&\cdots,&&x[i],&&\cdots,&&x[n-1]\big)&
\\
&\bm{x}_{i+1}&&:=\big(x[2],&\;&x[3],&\;&\cdots,&\;&x[i+1],&\;&\cdots,&\;&x[n]\big)&
\end{alignat}
$$

#### **》自己共分散関数**

特に，$\bm{x}_{i}$ と $\bm{x}_{i+1}$ の類似度を**共分散**で測りつつ，時間のずれを変数化した

$$
C_{xx}[k] = \sigma_{\bm{x}_{i}\bm{x}_{i+k}}
:=
\frac{1}{n-k}\sum_{i=1}^{n-k}(x[i]-m_x)(x[i+k]-m_x)
,\quad
m_x := \frac{1}{n}\sum_{i=1}^{n}x[i]
$$

のことを，**自己共分散関数 (autocovariance function, ACVF)** といいます．
<br/>時間間隔を表す変数 $k\geq 0$ を，**ラグ (lag)** といいます．

>自己共分散関数における平均中心化では，$\bm{x}_{i}$ や $\bm{x}_{i+1}$ の平均を使わず，元の時系列 $x[i]$ ($i=1,\cdots,n)$ の平均を使います.
>内積の定義からは若干外れますが，統計学においては，これが標準です．



#### >>>> 実装

````python
def ACVF(xs):
    '''
    自己共分散関数（DIY版）
    '''
    acvf = []
    length = len(xs)
    xmean = np.mean(xs)
    for lag in range(length-1):
        x1 = xs[0:length-lag]
        x2 = xs[lag:length]
        cov = np.mean((x1 - xmean) * (x2 - xmean))
        acvf.append(cov)

    return np.array(acvf)
````

#### **》自己相関関数**

また，相関係数と同じ発想で，自己共分散関数の値域を $-1$ から $1$ に正規化した

$$
R_{xx}[k] = \rho_{\bm{x}_{i}\bm{x}_{i+k}}
:=
\frac{\frac{1}{n-k}\sum_{i=1}^{n-k}(x[i]-m_x)(x[i+k]-m_x)}{\sigma_x^2}
=
\frac{C_{xx}[k]}{C_{xx}[0]}
,\quad
\sigma_x^2 := \frac{1}{n}\sum_{i=1}^{n}(x[i]-m_x)^2
$$

のことを，**自己相関関数 (autocorrelation function, ACF)** といいます．（$k\geq0$ はラグ）

> ここでも，平均 $m_x$ や分散 $\sigma_x^2$ は，元の時系列 $x[i]$ ($i=1,\cdots,n)$ のものです．

自己相関関数は，多くの場合，次のように使われます．

> **自己相関関数の使い方**
> - ラグ $k$ を横軸に，$R_{xx}[k]$ のグラフを描く．
> - もし，ラグ原点 $k_{p}=0$ にだけピークが存在し，それ以外で $R_{xx}[k]=0$ ならば，その時系列には，過去からの影響が認められない．すなわち，完全ランダムだとみなせる[<sup>*</sup>](#id_01)．
> - あるいは，特定のラグ間隔 $T$ でピークが繰り返すなら，その時系列は，周期 $T$ で似たパターンを繰り返している．
>
> <span id="id_01">* 厳密には，時系列長が有限だと，断言まではできません．</span>

> **別定義に関する補足**　信号処理や確率論では，平均中心化も正規化もしない次式を，自己相関関数と呼ぶことが多いです．
> $$
> R_{xx}[k]
> =
> \frac{1}{n-k}\sum_{i=1}^{n-k}x[i]x[i+k]
> \quad\left(
> =
> \frac{1}{n-k}\sum_{i=k+1}^{n}x[i-k]x[i]
> \right)
> $$

#### >>>> 実装

````python
def ACF(xs):
    '''
    自己相関関数（DIY版）
    '''
    acvf = ACVF(xs)

    return acvf/acvf[0]
````

#### **》動作確認**

````python
def graph_paper(lab1, lab2):
    """グラフ用紙の作成"""
    fig, ax = plt.subplots(2,1)
    ax[0].set_xlabel(r'time $i$')
    ax[0].set_ylabel(lab1)
    ax[0].grid();
    ax[1].set_xlabel(r'lag $k$')
    ax[1].set_ylabel(lab2)
    ax[1].grid(); fig.tight_layout()

    return fig, ax
````
